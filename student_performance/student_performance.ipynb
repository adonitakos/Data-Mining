```python
import numpy as np # matrix data structures
import matplotlib.pyplot as plt # data plotting
import pandas as pd # data analysis
from apyori import apriori # apriori algorithm for assocation rules & results
```

```python
# Read in the csv file into a dataframe-> 2D matrix used by pandas
student_data = pd.read_csv('exams.csv')
```

```python
student_data.head()
```

```python
record = []
for i in range(1,1000): # There are about 1000 rows in the dataset
    record.append([str(student_data.values[i,j]) for j in range(1,8)]) # There are 8 fields in the dataset
```

```python
# Apply Apriori to get the association rules
association_rules = apriori(record, min_support=0.01, min_confidence=0.4, min_lift = 2, min_length = 2)
# min_support <--- each set of scores, on a scale of 0-100 in this 1000 row data set, may only repeat themselves about 10 times
# min_confidence <--- I am confident through my own eye observations that scores amongst the three subjects are heavily correlated (i.e: Reading, Writing, Math consistently high and low amongst each student)
association_results = list(association_rules)
```

```python
print(len(association_results))
```
