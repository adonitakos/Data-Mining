```python
# Import dependencies
import pandas as pd
import matplotlib.pyplot as plt
```

```python
m1_rates = pd.read_csv('M1_rates.csv') # this existing datset was manually sorted
m1_rates.head()
```

```python
m1_tpr = m1_rates["TPR"]
m1_fpr = m1_rates["FPR"]

# Plot the ROC curve.
plt.plot(m1_fpr, m1_tpr)
plt.xlabel("False positive rate")
plt.ylabel("True positive rate")
plt.title("M1 ROC curve")
plt.show()
```
